## Variables
Variables are containers that hold values. They allow you to store and manipulate data in your program.

```python
balance = 1000
deposit_amount = 500
new_balance = balance + deposit_amount
print("New balance is: $" + str(new_balance))
```

This code uses variables to store the initial balance, the amount being deposited, and the new balance after the deposit. It then prints out the new balance.

## Conditionals
Conditionals are statements that allow your program to make decisions based on whether certain conditions are true or false.

```python
balance = 1000
withdrawal_amount = 100
if withdrawal_amount <= balance:
    balance = balance - withdrawal_amount
    print("New balance is: $" + str(balance))
else:
    print("Insufficient funds")
```

This code uses an `if` statement to check if the withdrawal amount is less than or equal to the balance. If it is, it subtracts the withdrawal amount from the balance and prints out the new balance. If it's not (`else`), it prints out "Insufficient funds".

## Loops
Loops allow your program to repeatedly execute a block of code until a certain condition is met. There are two main types of loops in Python:
- `for` loops
- `while` loops.

```python
balance = 1000
interest_rate = 0.05
num_years = 5
for i in range(num_years):
    balance = balance * (1 + interest_rate)
    print("Balance after year " + str(i+1) + " is: $" + str(balance))
```

This code uses a `for` loop to calculate the balance after 5 years of compound interest. It multiplies the balance by the interest rate plus 1, and prints out the new balance for each year.

Interest Rate is 5 percent (0.05).
If you have a 1000 dollars, 5% interest of this is $50 dollars.

## Functions
Functions are reusable blocks of code that perform a specific task. They allow you to break your code into smaller, more manageable pieces.

```python
def deposit(balance, amount):
    balance += amount
    return balance

def withdrawal(balance, amount):
    if amount <= balance:
        balance -= amount
        return balance
    else:
        print("Insufficient funds")
        return balance

balance = 1000
balance = deposit(balance, 500)
balance = withdrawal(balance, 200)
print("Current balance is: $" + str(balance))
```

This code defines two functions, deposit and withdrawal, which perform their respective actions and return the new balance. It then uses those functions to deposit $500 and withdraw $200 from an initial balance of $1000, and prints out the current balance.

## Comments
Comments are notes that you can add to your code to explain what it does or to remind yourself of something. They are ignored by the interpreter and are only meant for human readers


```python
# This program calculates compound interest on a bank account
balance = 1000.0
interest_rate = 0.05
num_years = 5

# Use a loop to calculate the balance after each year
for i in range(num_years):
    balance = balance * (1 + interest_rate)
    print("Balance after year " + str(i+1) + " is: $" + str(balance))
```

This code uses comments to explain what the program does and how it does it. It explains that the program calculates compound interest on a bank account, and adds comments to the variables and loop to explain what they do.

## Group By
`df.groupby(col_name)` organizes a DataFrame (df) into groups based on the unique values in one or more specified columns. It follows a "split-apply-combine" strategy, allowing you to perform operations on subsets of your data.   [Docs](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html)



> Ask yourself, *What am I looking to achieve with this data?* In this case below, I am looking to get the College Major demographics.

```python
# Get the demographics by major. There should be a count of students per major
student_count_by_major = df.groupby('Major').count()
student_count_by_major
```

```python
biology_majors_df = df[df['Major'] == 'Biology']
display(biology_majors_df)
```

## Visualizing Data

Another library named, Matplotlib allows creation of static, animated, and interactive visualizations in Python.

See examples [here](https://matplotlib.org/stable/gallery/index.html).

```python
# Plot the Major demographics using a pie chart
import matplotlib.pyplot as plt

# Create the pie chart
plt.figure(figsize=(10, 6)) # Optional: Adjust figure size
plt.pie(df.groupby('Major').count()['#'],
        labels=df.groupby('Major').count().index,
        shadow=False,
        autopct='%1.1f%%', # Format the percentage labels
        startangle=90) # Start the first slice at 90 degrees

plt.title("Sci+Tech Scholars 2026 Class Majors") # Add a title
plt.axis('equal') # Ensures the pie chart is drawn as a perfect circle
plt.show() # Display the plot
```
